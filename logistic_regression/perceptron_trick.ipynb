```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
plt.scatter(x=X[:,0],y=X[:,1],c=y)
plt.show()
```

<h2>Perceptron trick</h2>

```python
def perceptron(x,y):
    X = np.insert(x,0,1,axis=1)
    weights = np.ones(X.shape[1])
    learning_rate = 0.1

    for i in range(1000):
        random_row = np.random.randint(0,100)
        y_hat = step(np.dot(X[random_row],weights))
        weights = weights + learning_rate * (y[random_row]-y_hat)*X[random_row]

    return weights[0],weights[1:]

def step(z):
    return 1 if z>0 else 0
```

```python
_intercept,_coef = perceptron(X,y)
```

```python
print(_intercept)
print(_coef)
```

```python
m = -(_coef[0]/_coef[1])
b = -(_intercept/_coef[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='black',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
