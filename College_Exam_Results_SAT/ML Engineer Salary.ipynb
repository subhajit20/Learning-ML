<h1>Handling categorical data</h1>

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
```

```python
df['remote_ratio'].value_counts()
```

```python
# scl = StandardScaler()
# df['salary'] = scl.fit_transform(df[['salary']])
```

```python
y = df['salary']
x = df.drop(columns=['salary','work_year'])

X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2, random_state=44)
```

<h1>Creating machine learning pipeline</h1>

```python
X_train.sample(5)
```

```python
transformer = ColumnTransformer(
    transformers=[
        ('categorical', OneHotEncoder(), ['job_title','company_location']),
        ('ordinal', OrdinalEncoder(categories=[['EN','MI','SE','EX']]), ['experience_level']),
        ('ordinal1', OrdinalEncoder(categories=[['S','L','M']]), ['company_size'])
    ],
    remainder='passthrough'
)
```

```python
model = Pipeline(steps=[
    ('transformer', transformer),
    ('linear_regression', LinearRegression())
])
```

```python
model.fit(X_train,Y_train)
Y_pred = model.predict(X_test)
```

```python
r2_score(Y_test,Y_pred)
```
