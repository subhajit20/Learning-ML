<h1>Univeriate Analysis for Numarical Data</h1>

<h2>a. Histogram</h2>

```python
sns.histplot(data=df, x='Age', bins=80)
```

<h1>Univeriate Analysis for Categorical Data</h1>

```python
sns.countplot(x="Survived",data=df,legend=True)
plt.title('Univariate Analysis of Categorical Data')
plt.show()
```

```python
df["Survived"].value_counts().plot(kind="bar")
```

```python
plt.pie(df["Survived"].value_counts(),labels=df["Survived"].value_counts(),autopct='%.1f')
```

```python
df["Survived"].value_counts()
```

```python
df["Sex"].value_counts()
```

<h1>Biveriate Analysis</h1>

<h2>Biveriate Analysis between Categorical - Categorical column</h2>

```python
pd.crosstab(index=df["Pclass"],columns=[df["Survived"],df["Sex"]])
```

```python
sns.heatmap(pd.crosstab(df["Survived"],df["Sex"]))
```

```python
sns.heatmap(pd.crosstab(df["Sex"],df["Survived"]))
```

```python
sns.heatmap(pd.crosstab(index=[df["Survived"],df["Sex"]],columns=df["Pclass"]))
```

```python
sns.heatmap(pd.crosstab(index=df["Survived"],columns=[df["Pclass"],df["Sex"],df["Embarked"]]))
```

<h2>b.Biveriate Analysis between Numerical - Categorical column</h2>

```python
sns.barplot(x=df["Pclass"],y=df["Age"],hue=df["Survived"])
```

```python
sns.boxplot(x=df["Sex"],y=df["Age"],hue=df["Survived"])
```
